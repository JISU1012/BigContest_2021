```python
import pandas as pd
```

```python
long = pd.read_csv("./월별데이터 정리 (2)/month_long_term_frgn.csv")
long
```

```python
long['m'] = pd.to_datetime(long['m']) 
#long.info()
long.head(2)
```

```python
long['resd_ave'] = long['sum_resd_pop_cnt'] / long['count_num']
long['work_ave'] = long['sum_work_pop_cnt'] / long['count_num']
long['visit_ave'] = long['sum_visit_pop_cnt'] / long['count_num']
```

```python
#국가명 리스트
nation = list(long['nationality'].unique())
#동 리스트
dong = list(long['emd_nm'].unique())
dong
```

```python
# resd_ave dataframe
#윤서꺼 참조했어요 ,,ㅎ

dt_index = pd.date_range(start='20180101', end='20210601',freq='MS')
val1 = list(dt_index)*len(dong)
val2 = []
for i in dong:
    for j in range(0,len(dt_index)):
        val2.append(i)
        
resd_df = pd.DataFrame({'Year_Month':val1, 'emd_nm':val2})
resd_df
```

```python
resd_ave_list = list(np.array(long['resd_ave'].tolist()))
array_nat = np.zeros(shape=(len(resd_df),len(nation)))
```

```python
for i in range(0,len(resd_df)):
    date = resd_df['Year_Month'][i]
    emd = resd_df['emd_nm'][i]
    for j in range(0,len(nation)):
        sample_list = long[(long['m']==date)&(long['emd_nm']==emd)&(long['nationality']==nation[j])].index.tolist()
        if sample_list == []:
            array_nat[i,j] = np.NaN
        else:
            array_nat[i,j] = resd_ave_list[sample_list[0]]
```

```python
nat_df = pd.DataFrame(array_nat,columns=nation)
```

```python
data_df = pd.concat([resd_df,nat_df],axis=1)
data_df
```
